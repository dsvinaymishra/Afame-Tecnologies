# statistical checking of the significance columns

```python
import scipy.stats as stats
```

```python
# categorical columns
significant_cat_cols = []
# ho: that the predictors and target are independent . It means
# predictor variables are not helping in predicting the Target 

#ha: Ho is false


for i in categorical_columns:
    tbl = pd.crosstab(df_new.loc[:,i],df_new.Attrition)
    teststats,pvalue,deg,exp_freq = stats.chi2_contingency(tbl)
    if pvalue<0.05:
        significant_cat_cols.append(i)
    print(i,pvalue)
```
